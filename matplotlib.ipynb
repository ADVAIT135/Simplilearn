## Operator Description 
### fig.add_axes() -> Initializes subplot
### a = fig.add_subplot(2,2,2)
### fig, b = plt.subplots(nrows=3, ncols=2) -> Add Plots
### ax = plt.subplots(2,2) -> Create Subplot

```python
fig = plt.figure()
axis = fig.add_axes([0.5, 0.1, 0.8, 0.8]) #Control the left, right width , height of thr canvas (from 0 to 1)
axis.plot(x,y,'r')
```

```python
fig, axes = plt.subplots(nrows = 3, ncols=1)
for ax in axes:
    ax.plot(x,y,'r')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #Big Canvas
axes2 = fig.add_axes([0.2,0.5, 0.4, 0.3]) #Small Canvas

axes1.plot(x,y,'r')
axes2.plot(y,x,'g')
```

```python
fig = plt.figure(figsize=(10,9), dpi=350)
axis = fig.add_axes([0.5, 0.1, 0.8, 0.8]) #Control the left, right width , height of thr canvas (from 0 to 1)
axis.plot(x,y,'r')
```

```python
ax.legend(["Label 1", "Label 2"])
fig, axes = plt.subplots()
axes.set_xlabel("X-label")
axes.set_ylabel("Y-label")
axes.set_title("Title")
axes.plot(x,y, 'y')
```

## Other 2-D Graphs

```python
n = np.array([0,1,2,3,4,5])

fig, axes = plt.subplots(1,4, figsize=(16,5), dpi = 150)

axes[0].set_title("Scatter Plot")
axes[0].scatter(x, x +0.25*np.random.randn(len(x)))

axes[1].set_title("Step Plot")
axes[1].step(n, n**2, lw=2)

axes[2].set_title("Bar Plot")
axes[2].bar(n , n**2, align='center', width = 0.9, alpha = 0.9)

axes[3].set_title("Fill Between Plot")
axes[3].fill_between(x, x**2, x**3, color='gold', alpha=0.9)
```

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_axes([0.0, 0.0, 1, 1], polar=True)
t = np.linspace(0, 2*np.pi, 100)
ax.plot(t,t*3, color='red', lw=4)
```

```python
n = np.random.rand(1000)

fig, axes = plt.subplots(1,2, figsize=(12,4))

axes[0].set_title("Histogram - 1")
axes[0].hist(n)

axes[1].set_title("Cumulative Detailed Histogram - 1")
axes[1].hist(n, cumulative=True, bins=10)
```

## Contour Map

```python
import matplotlib
import numpy as mp
import matplotlib.cm as cm
import matplotlib.pyplot as plt


delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X, Y = np.meshgrid(x,y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1 )**2 - (Y-1)**2)
Z = (Z1 - Z2) * 2

print(X)
print(Y)
```

```python
fig, ax = plt.subplots(dpi = 200)
CS = ax.contour(X,Y,Z)
ax.clabel(CS, inline=1, fontsize=10)
ax.set_title("Contour Map")
```
