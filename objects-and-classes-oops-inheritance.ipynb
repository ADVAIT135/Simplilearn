## Objects and Classes

```python
class Person:
    def __init__(self):
        self.name = "Adu"
        self.gender = "male"
        self.age = 23
        
    def talk(self):
        print("Hi I am ", self.name, " and my gender is ", self.gender, " and I am ", self.age, "years old.")
        
    def vote(self):
        if self.age < 18:
            print("I am not eligible to vote")
        else:
            print("I am eligible to vote")

obj = Person()

Person.talk(obj)
Person.vote(obj)
print("\n\n")

obj.talk()
obj.vote()
```

```python
class Person:
    def __init__(self, n, g, a):
        self.name = n
        self.gender = g
        self.age = a
        
        # name, age and gender are attributes of Class Person 
        # talk() and you vote() are methods of Class Person
        
    def talk(self):
        print("Hi I am ", self.name, " and my gender is ", self.gender, " and I am ", self.age, "years old.")
        
    def vote(self):
        if self.age < 18:
            print("I am not eligible to vote")
        else:
            print("I am eligible to vote")
            
obj1 = Person("Sam", "Male", 17)  # obj1 is object of class Person
obj1.talk()
obj1.vote()
print("\n")

obj2 = Person("Jessie", "Female", 19) # obj2 is also object of class Person
obj2.talk()
obj2.vote()
print("\n")

print(obj1.name, obj2.name)
print(obj1.age, obj2.age)
print(obj1.gender, obj2.gender)
```

## Object Oriented Programming (OOPS)

```python
class car:        
    def get_speed(self):
        print("155 mph")
        
BMW = car()
Ford = car()

car.get_speed(BMW)
car.get_speed(Ford)
print("\n")
BMW.get_speed()
Ford.get_speed()
```

```python
class car(): 
    def __init__(self, year, speed):
        self.year = year
        self.speed = speed
    def get_speed(self):
        print("Maximum speed is ", self.speed)

BMW = car(2018, 155)
Ford = car(2016, 140)
print("\n")
BMW.get_speed()
Ford.get_speed()
```

```python
class car(): 
    def __init__(self, year, speed):
        self.year = year
        self.speed = speed
        
    def get_speed(self):
        print("Maximum speed is ", self.speed)
        
    def set_speed(self, speed):
        self.speed = speed

BMW = car(2018, 155)
BMW.set_speed(143)
BMW.get_speed()
print("\n")

Ford = car(2016, 140)
Ford.get_speed()
```

## Inheritance

```python
class car(): 
    def __init__(self, year, speed):
        self.year = year
        self.speed = speed
        
    def get_speed(self):
        print("Maximum speed is ", self.speed)
        
    def set_speed(self, speed):
        self.speed = speed

class Sedan(car): # Child Class
    def accelerate(self):
        print('137')
    def openTrunk(self):
        print('The Trunk has been opened.')
        
class SUV(car): # Child Class
    def accelerate(self):
        print('127')

Honda = Sedan(2018, 150)
BMW.get_speed()
Honda.get_speed()

Honda.openTrunk()
```

## Polymorphism 

```python
class Car: #parent Class
    def __init__(self, name):
        self.name = name
    
class Sedan(Car): #Child Class
    def accelerate(self):
        print(150)
    
class SUV(Car): #Child Class
    def accelerate(self):
        print(180)
    
ObjX = [Sedan("Camry"), SUV("Scorpio")]

for obj in ObjX:
    print(obj.name+" : ",end="")
    obj.accelerate()
```
