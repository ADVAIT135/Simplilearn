## Functions in Python

```python
def welcome():
    print("Good Morning")
```

```python
welcome()
```

```python
#Keyword Argument
def add(x, y):
    total = int(x)+int(y)
    print("The total of ", x, " and ", y, " is ", total)
```

```python
#Keyword Argument
add(y=3,x=4)
```

```python
#Default value
def add(x, y=10):
    total = int(x)+int(y)
    print("The total of ", x, " and ", y, " is ", total)

add(5)
```

```python
def add(*x):
    total = 0
    for i in x:
        total = total + i
    print("The total is ", total)

add(5, 10, 15 , 20, 78, 79 , 990)
```

```python
# Call by Reference
def add(x, y):
    x = 10
    y = 30
    print(id(x), id(y))
    total = int(x)+int(y)
    print("The total of ", x, " and ", y, " is ", total)
    
x = 40
y = 50
print(id(x), id(y))
add(x,y)
print("The sum is", x+y)
```

```python
def add(lst):
    lst[2] = 0
    
lst = [0,1,2]
print(lst)
add(lst)
print(lst)
```

```python
def add(x, y):
    total = int(x)+int(y)
    return total

result = add(10,20)    
print("The total is ", result)
```
