```python
from threading import *
def show():
    print("This is a child thread")

t = Thread(target=show)
t.start()

print("This is a parent thread")
```

```python
from threading import *
class Mythread(Thread):
    def run(self):
        for i in range(5):
            print("\nThis is a child class")
            
t = Mythread()
t.start()
for i in range(5):
    print("\nThis is the main thread")
```

```python
from threading import *
class Demo:
    def show(self):
        for i in range(5):
            print("This is the child thread")

obj = Demo()
t = Thread(target=obj.show())
t.start()

for i in range(5):
    print("This is the parent thread")
```

# MultiThreading

```python
from threading import *
import time
class Demo:
    def num(self):
        for i in range(1,6):
            print("The number is ", i)
            time.sleep(1)
    
    def double(self):
        for i in range(1,6):
            print("The double of the number", i, " is ", i*2)
            time.sleep(1)
            
    def square(self):
        for i in range(1,6):
            print("The square of the number", i, " is ", i*i)
            time.sleep(1)
    
obj = Demo()

t1 = Thread(target = obj.num)
t2 = Thread(target = obj.double)
t3 = Thread(target = obj.square)

t1.start()
time.sleep(0.2)
t2.start()
time.sleep(0.2)
t3.start()

t1.join()
t2.join()
t3.join()

print("This is the main thread")
```
